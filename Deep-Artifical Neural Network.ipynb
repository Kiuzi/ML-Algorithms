```python
dataset.head(10)
```

```python
# Encoding categorical data
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

# dummy for Gender
labelencoder_X = LabelEncoder()
X.iloc[:,2] = labelencoder_X.fit_transform(X.iloc[:,2])

#dummy for Country
dummy_GEO = pd.get_dummies(X["Geography"])
X.drop("Geography", axis = 1, inplace=True)
dummy_GEO.drop(columns=['Spain'], inplace=True)
X = pd.concat([X, dummy_GEO], axis=1)
X[0:5]
```

```python
X=X.values
X[0:5]
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
import tensorflow as tf
```

```python
from tensorflow import keras
```

```python
from tensorflow.keras import layers
```

```python
# Importing the Keras libraries and packages

#Firstly, you should download and set up the anaconda via anaconda and then open anaconda prompt and then
#follow these steps via TensorFlow. Fourthly, you should use 'pip install --ignore-installed --upgrade tensorflow==1.14.0'
#to install TensorFlow 1.14 on Conda Finally, you can use TensorFlow that is version 1.14 on python 3.7
# OR visit this website:https://www.tensorflow.org/install/pip
#from keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Initialising the ANN
classifier = tf.keras.Sequential()
```

```python
# Adding the input layer and the first hidden layer

classifier.add(layers.Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11))
```

```python
# Adding the second hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
```

```python
# Adding the output layer
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

```python
# Compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# Fitting the ANN to the Training set
classifier.fit(X_train, y_train, batch_size = 10, nb_epoch = 100)
```

```python
# Part 3 - Making the predictions and evaluating the model

# Predicting the Test set results
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
y_pred[0:6]
```

```python

# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
# Encoding categorical data
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

# dummy for Gender
labelencoder_X = LabelEncoder()
X.iloc[:,2] = labelencoder_X.fit_transform(X.iloc[:,2])

#dummy for Country
dummy_GEO = pd.get_dummies(X["Geography"])
X.drop("Geography", axis = 1, inplace=True)
dummy_GEO.drop(columns=['Spain'], inplace=True)
X = pd.concat([X['CreditScore'], dummy_GEO,X.iloc[:,1:]], axis=1)
X[0:5]
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Importing the Keras libraries and packages

# conda install pip
# pip3 install --upgrade tensorflow #Change pip3 to pip if did not work
# pip install Keras
```

```python
import tensorflow as tf 
```

```python
from tensorflow import keras
```

```python
from tensorflow.keras import layers
```

```python
from keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense
```

```python
# Initialising the ANN
classifier = tf.keras.Sequential()
```

```python
# Adding the input layer and the first hidden layer

classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11))
```

```python
# Adding the second hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
```

```python
# Adding the output layer
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

```python
# Compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# Fitting the ANN to the Training set
classifier.fit(X_train, y_train, batch_size = 16, nb_epoch = 100)
```

```python
# Part 3 - Making the predictions and evaluating the model

# Predicting the Test set results
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
y_pred[0:6]
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```
